```python
import pandas as pd
```

```python
df=pd.read_csv("Dataset/Superstore.csv",encoding='latin-1')
print(df.head)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df["Order Date"]
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month_name()
df["Month Number"] = df["Order Date"].dt.month
```

```python
df.describe()
```

```python
print(df["Category"].unique())
print(df["Region"].unique())
print(df["Segment"].unique())
print(df["Ship Mode"].unique())
```

```python
df.to_csv("Dataset/Cleaned.csv", index=False)
```

```python
sales_year = df.groupby("Year")["Sales"].sum()
print(sales_year)
```

```python
sales_month = df.groupby("Month")["Sales"].sum()
print(sales_month)
```

```python
sales_category=df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print(sales_category)
profit_category=df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
print(profit_category)
```

```python
sales_region=(df.groupby("Region")["Sales"].sum().sort_values(ascending=False))
print(sales_region)
profit_region=(df.groupby("Region")["Profit"].sum().sort_values(ascending=False))
print(profit_region)
```

```python
top_products=(df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10))
print(top_products)
```
